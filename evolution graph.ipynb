```
import pandas as pd
import pylab
import matplotlib.pyplot as plt
# make the plots a little wider by default
%matplotlib inline
plt.style.use('ggplot')

pylab.rcParams['figure.figsize'] = (10., 15.)

# Load the historical election data (1996-2020)
historical_data_url = "https://raw.githubusercontent.com/Trisken221/QM2/refs/heads/main/1996-2020%20.csv"
historical_data = pd.read_csv(historical_data_url)
historical_data
```

```
historical_data.rename(columns={'Unnamed: 0': 'State'}, inplace=True)
historical_data.dropna(subset=['State'], inplace=True)
columns = historical_data.columns.tolist()
id_vars = ['State']
value_vars = [col for col in columns if col not in id_vars]
historical_data_long = historical_data.melt(id_vars=id_vars, value_vars=value_vars, var_name='Year_Party', value_name='Vote_Percentage')
```

```
historical_data_long[['Year', 'Party']] = historical_data_long['Year_Party'].str.extract(r'(\d{4})\s(.*)')
historical_data_long["Vote_Percentage"] = pd.to_numeric(historical_data_long["Vote_Percentage"].astype(str).str.replace('%', ''), errors='coerce')
historical_data_long = historical_data_long[historical_data_long['Party'].isin(['Democratic', 'Republican'])]
```

```
#clean and integrate with 2024 data
data = pd.read_csv('https://raw.githubusercontent.com/Trisken221/QM2/refs/heads/main/2024%20election%20results%20by%206th%20December.csv')
data
```

```
selected_data = data[['State', 'Party', 'Vote share']]

# Clean and prepare the 2024 data
selected_data.rename(columns={"Vote share": "Vote_Percentage"}, inplace=True)

selected_data["Vote_Percentage"] = pd.to_numeric(selected_data["Vote_Percentage"].str.replace('%', '', regex=False), errors='coerce')
selected_data = selected_data[selected_data['Party'].isin(['D', 'R'])]
selected_data["Party"] = selected_data["Party"].apply(lambda x: "Democratic" if x == "D" else "Republican")
selected_data = selected_data.dropna(subset=["Vote_Percentage"])
selected_data
```

```
# Group by State and find the party with the highest vote percentage
winning_party = selected_data.groupby('State').apply(lambda df: df.loc[df['Vote_Percentage'].idxmax()])
winning_party.reset_index(drop=True, inplace=True)

# Now winning_party DataFrame contains the state and the party with the highest vote percentage
print(winning_party[['State', 'Party', 'Vote_Percentage']])
```

```
# Reshape the DataFrame
historical_data_long = historical_data.melt(
    id_vars=["State"], var_name="Year_Party", value_name="Vote_Percentage"
)

historical_data_long[['Year', 'Party']] = historical_data_long['Year_Party'].str.extract(r'(\d{4})_(.*)')

# Convert 'Year' to numeric
historical_data_long["Year"] = pd.to_numeric(historical_data_long["Year"], errors='coerce')
historical_data['Vote_Percentage'] = pd.to_numeric(historical_data['Vote_Percentage'], errors='coerce')

# Determine the winning party based on vote percentage
historical_data_long["Party_Held"] = historical_data_long.apply(
    lambda row: "Democratic" if row["Vote_Percentage"] > 50 else "Republican" if row["Vote_Percentage"] <= 50 else "Tie",
    axis=1
)

# Filter out rows with vote percentage less than 50% or those classified as 'Tie'
historical_data_long = historical_data_long[
    (historical_data_long["Party_Held"] == "Democratic") | (historical_data_long["Party_Held"] == "Republican")
]

# Display the reshaped DataFrame
print(historical_data_long)
```

```
# Standardize and rename columns in historical_data
historical_data.columns = historical_data.columns.str.strip()  # Remove leading/trailing spaces
historical_data

# Verify the presence of the 'State' column
print("Columns in historical_data:", historical_data.columns)

historical_data = historical_data.melt(
    id_vars=["State"], var_name="Year_Party", value_name="Vote_Percentage"
)
historical_data[["Year", "Party"]] = historical_data["Year_Party"].str.extract(r"(\d{4})_(.*)")
historical_data["Vote_Percentage"] = pd.to_numeric(historical_data["Vote_Percentage"], errors='coerce')

# Determine the winning party for each state and year
def determine_winning_party(row):
    if row["D"] > row["R"]:
        return "Democrat"
    elif row["R"] > row["D"]:
        return "Republican"
    else:
        return "Tie"

historical_data_pivot = historical_data.pivot_table(
    index=["State", "Year"], columns="Party", values="Vote_Percentage"
).reset_index()
historical_data_pivot["Party_Held"] = historical_data_pivot.apply(determine_winning_party, axis=1)
historical_data_pivot = historical_data_pivot[["Year", "State", "Party_Held"]]

# Combine historical data and 2024 data
final_data = pd.concat([historical_data_pivot, selected_data], ignore_index=True)

# Sort and save the merged data
final_data = final_data.sort_values(by=["Year", "State"]).reset_index(drop=True)
final_data.to_csv("merged_election_data.csv", index=False)

# Display the final merged dataset
print(final_data.head())
```
