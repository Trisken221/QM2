```
import pandas as pd
import pylab
import matplotlib.pyplot as plt
# make the plots a little wider by default
%matplotlib inline
plt.style.use('ggplot')

pylab.rcParams['figure.figsize'] = (10., 15.)

# Load the historical election data (1996-2020)
historical_data_url = "https://raw.githubusercontent.com/Trisken221/QM2/refs/heads/main/1996-2020%20election%20results.csv"
historical_data = pd.read_csv(historical_data_url, na_values=['', ' '])

from IPython.display import display
display(historical_data)
```

```
historical_data.fillna('', inplace=True)
historical_data.columns = [col if 'Unnamed' not in col else '' for col in historical_data.columns]
historical_data = historical_data.loc[~(historical_data == '').all(axis=1)] # Fixed indentation

display(historical_data)
```
